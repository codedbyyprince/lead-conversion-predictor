```python
nunique_df = []
for i in df.columns:
    nunique_df.append(f'{i} : {df[i].nunique()}')
    
cols = []
for item in nunique_df:
    col, val = item.rsplit(" : ", 1)
    val = int(val)
    
    if 2 <= val <= 200:
        cols.append(col)

for i in cols:
    print('------------------------')
    print(df[i].value_counts(dropna=False))
    print('------------------------')
    print()
```

```python
# All of these are either completely constant (1 unique value)
# or have 9,230+ rows on one side — no signal for any model

zero_var_cols = [
    "What matters most to you in choosing a course",
    "Magazine",
    "Receive More Updates About Our Courses",
    "Update me on Supply Chain Content",
    "Get updates on DM Content",
    "I agree to pay the amount through cheque",
    "Do Not Call",
    "Newspaper Article",
    "X Education Forums",
    "Newspaper",
    "Digital Advertisement",
    "Through Recommendations",
    "Search",
]

df.drop(columns=zero_var_cols, inplace=True)
```

```python
# Asymmetrique Index cols: 46% null, redundant with the Score cols
# dropign values non usefull and highly null

df.drop(columns=[
    "Lead Quality",
    "Lead Profile",
    "Asymmetrique Activity Index",
    "Asymmetrique Profile Index",
    "How did you hear about X Education",
], inplace=True)
```

```python
df.info()
```

```python
# "Select" is a dropdown default — same as leaving it blank
# Convert to NaN so the imputation step handles it uniformly

placeholder_cols = ["Specialization", "Lead Profile", "City"]

for col in placeholder_cols:
    if col in df.columns:
        df[col] = df[col].replace("Select", np.nan)
```

```python
# TotalVisits and Page Views Per Visit: 137 nulls each

df['TotalVisits'] = df['TotalVisits'].fillna(df['TotalVisits'].median())
df['Page Views Per Visit'] = df['Page Views Per Visit'].fillna(df['Page Views Per Visit'].median())
```

```python
# Columns with small null % → fill with "Unknown" (honest label)
# Columns where "Select" was already replaced → use "Not Specified"

df["Last Activity"].fillna("Unknown", inplace=True)
df["What is your current occupation"].fillna("Unknown", inplace=True)

df["Specialization"].fillna("Not Specified", inplace=True)
# df["Lead Profile"].fillna("Not Specified", inplace=True)
# df["City"].fillna("Not Specified", inplace=True)
```

```python
nunique_df = []
for i in df.columns:
    nunique_df.append(f'{i} : {df[i].nunique()}')
    
cols = []
for item in nunique_df:
    col, val = item.rsplit(" : ", 1)
    val = int(val)
    
    if 2 <= val <= 200:
        cols.append(col)

for i in cols:
    print('------------------------')
    print(df[i].value_counts(dropna=False))
    print('------------------------')
    print()
```

```python
# Only 137 nulls = 1.5% which is under 20% rule so fill with median
df['TotalVisits'] = df['TotalVisits'].fillna(df['TotalVisits'].median())

# Anything above 20 becomes 21 — acts as a "heavy visitor" flag
df['TotalVisits'] = df['TotalVisits'].apply(lambda x: 21 if x > 20 else x)
```

```python
# Only 137 nulls = 1.5% which is under 20% rule so fill with median
# 114 unique float values is too many — round to nearest whole number
# This reduces unique values drastically and makes it cleaner
df['Page Views Per Visit'] = df['Page Views Per Visit'].round(0).astype(int)

df['Page Views Per Visit'] = df['Page Views Per Visit'].apply(
    lambda x: 11 if x > 10 else x
)

# Verify the result
print(df['Page Views Per Visit'].value_counts().sort_index())

df['Page Views Per Visit'] = df['Page Views Per Visit'].fillna(df['Page Views Per Visit'].median())
```

```python
# Only 36 nulls = 0.4% so fill with mode (most frequent value)
df['Lead Source'] = df['Lead Source'].fillna(df['Lead Source'].mode()[0])

# Standardise casing first so "google" and "Google" become the same
df['Lead Source'] = df['Lead Source'].str.strip().str.title()

# Find all categories that have less than 55 rows
small_sources = df['Lead Source'].value_counts()
small_sources = small_sources[small_sources < 55].index.tolist()

# Merge them all into "Other"
df['Lead Source'] = df['Lead Source'].apply(
    lambda x: 'Other' if x in small_sources else x
)
```

```python
# No NaN fill needed — already handled as "Unknown" by you earlier
# Find all categories with less than 30 rows
small_activities = df['Last Activity'].value_counts()
small_activities = small_activities[small_activities < 30].index.tolist()

# Merge them into "Other"
df['Last Activity'] = df['Last Activity'].apply(
    lambda x: 'Other' if x in small_activities else x
)
```

```python
# 96% of known values are India so one binary column carries all the signal
# Fill NaN as "Unknown" first so they don't cause issues
df['Country'] = df['Country'].fillna('Unknown')

# Create binary column — Yes if India, No for everything else
# Keeping as Yes/No text so it stays consistent with other binary columns
# Pipeline will handle the encoding later
df['is_india'] = df['Country'].apply(
    lambda x: 'Yes' if x.lower() == 'india' else 'No'
)

# Drop the original Country column, we don't need it anymore
df.drop(columns=['Country'], inplace=True)
```

```python
# Find categories with 16 rows or less
small_occ = df['What is your current occupation'].value_counts()
small_occ = small_occ[small_occ <= 16].index.tolist()

# Merge into "Other"
df['What is your current occupation'] = df['What is your current occupation'].apply(
    lambda x: 'Other' if x in small_occ else x
)
```

```python
# 36% nulls — above 20% rule so we do NOT fill with stats
# Instead just label them honestly as "Unknown"
df['Tags'] = df['Tags'].fillna('Unknown')

# Now merge all small categories under 55 rows into "Other"
small_tags = df['Tags'].value_counts()
small_tags = small_tags[small_tags < 55].index.tolist()

df['Tags'] = df['Tags'].apply(
    lambda x: 'Other' if x in small_tags else x
)
```

```python
# No NaN in this column so just merge small categories
small_notable = df['Last Notable Activity'].value_counts()
small_notable = small_notable[small_notable < 55].index.tolist()

df['Last Notable Activity'] = df['Last Notable Activity'].apply(
    lambda x: 'Other' if x in small_notable else x
)
```

```python
# Rural and Agribusiness (73), E-Business (57), Services Excellence (40)
# All under 100 and below the threshold — merging into Other
small_spec = ['Rural and Agribusiness', 'E-Business', 'Services Excellence']

df['Specialization'] = df['Specialization'].apply(
    lambda x: 'Other' if x in small_spec else x
)

# Verify
print(df['Specialization'].value_counts())
```

```python
# Fill NaN as Unknown — only 40% missing, honest label
df['City'] = df['City'].fillna('Unknown')

# Simplify categories to reduce noise
# "Other Cities" and "Other Cities of Maharashtra" stay as is — they are descriptive
# "Other Metro Cities" rename to "Metro Cities" — cleaner label
# "Tier II Cities" stays as is
# Everything else stays as is

df['City'] = df['City'].replace({
    'Other Metro Cities': 'Metro Cities'
})

# Verify
print(df['City'].value_counts())
```

```python
nunique_df = []
for i in df.columns:
    nunique_df.append(f'{i} : {df[i].nunique()}')
    
cols = []
for item in nunique_df:
    col, val = item.rsplit(" : ", 1)
    val = int(val)
    
    if 2 <= val <= 200:
        cols.append(col)

for i in cols:
    print('------------------------')
    print(df[i].value_counts(dropna=False))
    print('------------------------')
    print()
```
